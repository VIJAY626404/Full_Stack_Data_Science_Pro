 # 1. Define a doubly linked list  [ Will be done in the class ]

```python
class Node:
    def __init__(self,data,next=None,prev=None):      
        self.__data =data
        self.__prev = prev
        self.__next = next
        
    def getData(self):
        return self.__data
    def setData(self,data):
        self.__data =data
    
    # get & set for prev node
    def setPrev(self,prev):
        self.__prev = prev
    def getPrev(self):
        return self.__prev
    
    # get and set for next node
    def setNext(self,next):
        self.__next = next
    def getNext(self):
        return self.__next
    
head = Node(2)
node2 = Node(4)
node3 = Node(6)
node4 = Node(8)
node5 = Node(5)

# create link between nodes
head.setNext(node2)
node2.setNext(node3)
node2.setPrev(head)
node3.setNext(node4)
node3.setPrev(node2)
node4.setNext(node5)
node4.setPrev(node3)
node5.setPrev(node4)

def traverse(head):
    temp = head
    while(temp):
        print(temp.getData(),end='<->')
        temp  = temp.getNext()
    print(None)
    
print('Double Linked List: ')
traverse(head)
```

 # 2. Write a function to reverse a linked list in-place

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next =next

# printlist 
def printList(head):
    temp = head
    while(temp):
        print(temp.data,end='->')
        temp  = temp.next
    print(None)

# define function for reverse List
def reverse_linked_list(head):
    prev = None
    current = head
    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev
 
# define linked list
head = Node(2)
node2 = Node(4)
node3 = Node(6)
node4 = Node(8)
node5 = Node(10)

# create link between nodes
head.next= (node2)
node2.next= (node3)
node3.next= (node4)
node4.next= (node5)


print("Before Reverse List")
printList(head)
print('After Reverse List')
head = reverse_linked_list(head)
printList(head)
```

# 3.  Detect cycle in a linked list

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next =next

# defien function for detect cycle
def Detect_cycle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if fast and slow.data == fast.data:
            return True
    return False

 
# define linked list
head = Node(2)
node2 = Node(4)
node3 = Node(6)
node4 = Node(8)
node5 = Node(10)

# create link between nodes
head.next= (node2)
node2.next= (node3)
node3.next= (node4)
node4.next= (node5)
node5.next = node3


print('Is Cycle:',Detect_cycle(head))
```

# 4. Merge two sorted linked list into one
 1->3->5->6->null and 2->4->6->8->null should be merged to make
 1->2->3->4->5->6->7->8

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next =next

def merge_two_lists(l1, l2):
    
    dummy = Node()
    current = dummy

    while l1 and l2:
        if l1.data < l2.data:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next

    # If any list is not fully traversed, append the remaining nodes
    if l1:
        current.next = l1
    if l2:
        current.next = l2

    return dummy.next


def printList(head):
    temp = head
    while(temp):
        print(temp.data,end='->')
        temp  = temp.next
    print(None)
    

head1 = Node(1)
node1 = Node(3)
node2 = Node(5)
node3 = Node(7)

#create linked
head1.next = node1
node1.next = node2
node2.next = node3



head2 = Node(2)
node4 = Node(4)
node5 = Node(6)
node6 = Node(8)


head2.next = node4
node4.next = node5
node5.next = node6

print('First List is:')
printList(head1)
print('Second List is: ')
printList(head2)

newHead = merge_two_lists(head1,head2)
print('After Merge List:')
printList(newHead)
```

# 5.  Write a function to remove nth node from the end in a linked list

 #  6. Remove duplicates from a sorted linked list
 1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5

```python
class Node:
    def __init__(self,value=0,next = None):
        self.value = value
        self.next = next
    

def RemoveDuplicate(head):
    if not head or not head.next:
        return head
    current = head
    while current and current.next:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next
    return head

def printList(head):
    temp = head
    while(temp):
        print(temp.value,end='->')
        temp  = temp.next
    print(None)
    
        

# define node
head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(3)
node4 = Node(4)
node5 = Node(4)
node6 = Node(4)
node6 = Node(5)


# create linked between list
head.next = node1
node1.next = node2
node2.next = node3
node3.next  =node4
node4.next = node5
node5.next = node6

print('Origional List: ')
printList(head)
headA = RemoveDuplicate(head)
print("After Removing Duplicate Values:")
printList(headA)
```

 # 8. Rotate a linked list by k positions to the right
 1->2->3->4->8->6->9 , after rotating for 2 times becomes , 3->4->8->6->9->1->2

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head, k):
    if not head or k == 0:
        return head
    
    # Find the length of the linked list and locate the tail node
    length = 0
    tail = head
    while tail.next:
        tail = tail.next
        length += 1
    
    # Adjust k to be within the length of the list
    k = k % length
    
    if k == 0:
        return head
    
    # Connect the tail to the head to form a loop
    tail.next = head
    
    # Traverse to the (length - k)th node
    new_tail_position = length - k
    new_tail = head
    for _ in range(new_tail_position - 1):
        new_tail = new_tail.next
    
    # Update the head to the (length - k)th node's next
    new_head = new_tail.next
    
    # Break the loop by setting the new tail's next to None
    new_tail.next = None
    
    return new_head

# Helper function to print the linked list
def printList(node):
    while node:
        print(node.val, end=" -> ")
        node = node.next
    print("None")

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Original linked list:")
printList(head)

k = 2
new_head = rotateRight(head, k)

print(f"\nLinked list after merging {k} positions to the right:")
printList(new_head)
```

# 9. Add Two Numbers Represented by LinkedLists:
Given two non-empty linked lists representing two non-negative integers, where the digits are stored in 
reverse order, add the two numbers and return it as a linked list

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy_head = ListNode()  # Dummy head to simplify the code
    current = dummy_head  # Pointer to traverse the result list

    carry = 0  # Initialize the carry to 0

    while l1 or l2 or carry:
        # Calculate the sum of current digits and the carry
        sum_val = carry
        if l1:
            sum_val += l1.val
            l1 = l1.next
        if l2:
            sum_val += l2.val
            l2 = l2.next

        # Update the carry
        carry = sum_val // 10

        # Create a new node with the digit (sum % 10)
        current.next = ListNode(sum_val % 10)
        current = current.next

    return dummy_head.next  # Return the result linked list, excluding the dummy head

# Helper function to print the linked list
def printList(node):
    while node:
        print(node.val, end=" -> ")
        node = node.next
    print("None")

# Example usage:
# Create the first linked list: 2 -> 4 -> 3 (represents 342)
l1 = ListNode(2)
l1.next = ListNode(4)
l1.next.next = ListNode(3)

# Create the second linked list: 5 -> 6 -> 4 (represents 465)
l2 = ListNode(5)
l2.next = ListNode(6)
l2.next.next = ListNode(4)

# Add the two numbers represented by the linked lists
result = addTwoNumbers(l1, l2)

# Print the result linked list
print("Result:")
printList(result)
```

# 10.  Clone a Linked List with next and Random Pointer
 Given a linked list of size N where each node has two links: one pointer points to the next node and the 
second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time. 
Note: The pointer pointing to the next node is ‘next‘ pointer and the one pointing to an arbitrary node is 
called ‘arbit’ pointer as it can point to any arbitrary node in the linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def clone_linked_list(head):
    if not head:
        return None
    
    # Step 1: Create new nodes and insert them after original nodes
    current = head
    while current:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next
    
    # Step 2: Set next pointers of newly created nodes
    current = head
    while current:
        if current.next:
            current.next.random = current.random.next if current.random else None
        current = current.next.next
    
    # Step 3: Separate the cloned list from the original list
    original_head = head
    clone_head = head.next
    clone_current = clone_head
    
    while original_head:
        original_head.next = original_head.next.next
        clone_current.next = clone_current.next.next if clone_current.next else None
        
        original_head = original_head.next
        clone_current = clone_current.next
    
    return clone_head

def printList(node):
    while node:
        print(node.data, end=" -> ")
        node = node.next
    print("None")


# Create the original linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)

# Set random pointers
head.random = head.next.next  # 1 -> 3
head.next.random = head  # 2 -> 1
head.next.next.random = head.next.next.next  # 3 -> 4
head.next.next.next.random = head.next  # 4 -> 2

print('Linked List with next Pointer:')
printList(head)

# Clone the linked list
cloned_head = clone_linked_list(head)
print('Linked List that Pointing Random Pointer: ')

# Print the cloned linked list
while cloned_head:
    print("Data:", cloned_head.data)
    print("Random:", cloned_head.random.data if cloned_head.random else None)
    print("--------")
    cloned_head = cloned_head.next
```
