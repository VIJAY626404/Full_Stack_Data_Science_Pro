# <span style="color:blue; font-size:30px"> 1.Explain the concept of a loss function in the context of deep learning. Why are loss functions important in training neural networks?</span>

In deep learning, a **loss function** (also known as a cost function or objective function) is a mathematical function that measures how well a neural network model’s predictions align with the actual target values. It quantifies the "error" between the predicted output from the model and the actual output (the ground truth), providing a measure that the model aims to minimize during training.

## Purpose and Importance of Loss Functions
1. **Guiding Model Optimization:** The primary purpose of a loss function is to guide the model’s training process. During training, the model makes predictions, and the loss function calculates the error in these predictions. This error is then used in backpropagation, where the gradients of the loss with respect to each parameter are computed, allowing an optimization algorithm (like gradient descent) to update the weights to minimize the loss. By continuously updating the weights to reduce the loss, the model iteratively improves its predictions.

2. **Providing a Performance Metric:** Loss functions provide a numeric score representing the model's performance. By monitoring the loss value, one can gauge how well the model is learning over time. Lower loss values indicate better performance, while higher values suggest the model needs further training or adjustments to the learning process.

## Different Loss Functions for Different Tasks: 
The choice of the loss function depends on the type of machine learning task:

- **For regression tasks**, a common loss function is Mean Squared Error (MSE), which penalizes larger errors more heavily.
- **For classification task**s, especially multi-class, Cross-Entropy Loss is often used as it handles probabilities and penalizes the model for confident, incorrect predictions.
- **For binary classification**, Binary Cross-Entropy is a preferred choice, as it measures the accuracy in predicting probabilities of binary classes.

## Why Loss Functions are Essential
Loss functions are essential for:
- Guiding learning by providing feedback on how the model should adjust its parameters.
- Stabilizing training by providing a continuous value to minimize, ensuring convergence to a minimum.
- Enhancing accuracy by helping the model to focus on minimizing errors and improve prediction accuracy over time.

# <span style="color:blue; font-size:30px"> 3.Discuss the challenges associated with selecting an appropriate loss function for a given deep learning task. How might the choice of loss function affect the training process and model performance?</span>

**Selecting an appropriate loss function** for a given deep learning task can be challenging, and the choice has a significant impact on the training process, model convergence, and overall performance. The loss function is integral to guiding the model's learning behavior, so choosing the wrong one can lead to poor model performance, overfitting, underfitting, or unstable training. 

# Here are some key challenges and considerations in selecting a loss function:
### 1. Task Suitability and Problem Type
- **Matching Loss Function to Task:** Certain loss functions are more appropriate for specific types of tasks (e.g., Mean Squared Error for regression, - -Binary Cross-Entropy for binary classification). Using an incompatible loss function can misguide the model, leading to high error rates or suboptimal predictions. For example, applying MSE to a binary classification problem would fail to capture the probabilistic nature of binary outcomes and hinder performance.
- **Multi-Class vs. Multi-Label Classification:** In multi-class classification, each sample belongs to only one class (using Categorical Cross-Entropy), whereas in multi-label classification, a sample can belong to multiple classes simultaneously (typically requiring Binary Cross-Entropy with Sigmoid activation for each class). Misclassifying these tasks can lead to erroneous probabilities and poor generalization.

### 2. Sensitivity to Outliers
- Some loss functions, like Mean Squared Error (MSE), are highly sensitive to outliers, as the error is squared. This sensitivity can lead to significant weight updates for a few extreme points, potentially making the model focus disproportionately on outliers rather than generalizing well across the data.
- Alternatives like Mean Absolute Error (MAE) can reduce sensitivity to outliers by penalizing errors linearly rather than quadratically. However, MAE can be harder to optimize in some cases, as it lacks a smooth gradient at zero.
### 3. Class Imbalance
- In classification tasks with imbalanced data, standard loss functions like Binary Cross-Entropy or Categorical Cross-Entropy may perform poorly as they assume equal importance across classes. For instance, in a dataset with 95% of instances in one class, the model may achieve high accuracy by only predicting the majority class, while completely ignoring the minority class.
- To address this, variations such as Weighted Cross-Entropy (which assigns higher weight to the minority class) or Focal Loss (which penalizes hard-to-classify examples more heavily) can help, though tuning these weights is often a trial-and-error process.
### 4. Convergence and Gradient Behavior
- he choice of loss function affects how gradients behave during backpropagation. For example, Cross-Entropy produces strong gradients when predictions are far from actual labels, which can help speed up convergence. However, it can lead to unstable training if gradients are too high, especially in deep or complex networks.
- Loss functions like Huber Loss offer a compromise by behaving like MSE for small errors and like MAE for large errors, which can stabilize convergence in certain regression tasks. However, deciding the threshold (delta) for this behavior requires careful tuning.
### 5. Interpretability and Model Output Requirements
- Some loss functions are more interpretable or meaningful for specific tasks. For example, MSE provides error in terms of squared units, which may be intuitive for continuous outcomes, while Binary Cross-Entropy relates closely to probabilistic interpretation, useful in binary and multi-class classification.
- For tasks requiring specific outputs (e.g., probabilities for classification tasks), loss functions must align with the chosen activation function (e.g., softmax for Categorical Cross-Entropy). Using incompatible activation and loss function combinations (e.g., softmax with MSE) can cause poor performance, as gradients may not behave as expected.
### 6. Generalization and Overfitting
- Some loss functions can inadvertently encourage overfitting if they focus too heavily on reducing training error without promoting generalization. For instance, Mean Squared Error may lead to overfitting in regression tasks if the model becomes too focused on fitting every point, including noise.
- Regularization methods like L2 or L1 regularization or specific loss functions that promote sparsity or robustness (e.g., Huber Loss) can mitigate overfitting, though this adds complexity in selecting the regularization strength.
### 7. Computational Efficiency
- Certain loss functions are computationally more intensive than others, especially those that require complex computations or are sensitive to large datasets. For example, Contrastive Loss used in metric learning or Triplet Loss for face recognition tasks can be costly as they involve pairwise or triplet comparisons, which can scale poorly with large datasets.
- For efficiency in large-scale problems, simpler loss functions or sampling strategies (e.g., negative sampling in contrastive learning) are sometimes preferable to balance computational cost and model performance.
  
# Summary of Effects on Training and Performance
1. **Model Convergence:** Loss functions affect the speed and stability of convergence. Losses with sharp gradients (e.g., Cross-Entropy) may converge faster but risk instability, whereas smoother losses (e.g., MAE) provide stability but may converge more slowly.
2. **Performance and Accuracy:** The loss function directly impacts model accuracy. Incorrectly chosen loss functions can lead to suboptimal performance on the test set, particularly if they don’t match the task requirements or if they lead to imbalanced weight updates.
3. **Generalization:** The choice of loss function influences how well the model generalizes to new data. Loss functions that penalize outliers too heavily can lead to overfitting, whereas robust losses help the model generalize better by ignoring noise.
# Final Thought
Selecting a loss function is more than just aligning with the task type; it involves considering aspects like data distribution, class balance, interpretability, and computational resources. Experimenting with and tuning loss functions based on the problem requirements and dataset characteristics is often key to optimizing model performance.

**Weighted loss functions** in deep learning assign different levels of importance to certain samples or classes during training. By applying weights, these functions can help the model prioritize specific data points or categories, addressing issues like class imbalance or varying importance of errors across classes.

### When and Why to Use Weighted Loss Functions:
Weighted loss functions are useful when:
1. **Class Imbalance:** When one class is much more common than another (e.g., fraud detection, rare disease diagnosis), the model can become biased toward the majority class. Weighting the loss function helps give more importance to the minority class, encouraging the model to learn its characteristics.
2. **Importance-Based Prioritization:** When certain classes or errors carry more significance. For example, in medical diagnosis, misclassifying a severe disease may be more costly than misclassifying a mild one, so higher weights are assigned to severe cases.
3. **Noisy Data or Outliers:** Some samples may be noisy or unreliable. By assigning lower weights to these data points, the model can focus on more reliable samples, improving robustness.
### Examples of Scenarios Benefiting from Weighted Loss Functions:
1. **Image Classification with Imbalanced Classes:** In a dataset with many "cat" images but few "tiger" images, a weighted loss function like Weighted Cross-Entropy can reduce bias towards "cat" predictions, improving the accuracy for "tiger."
2. **Medical Imaging for Disease Detection:** In detecting rare diseases, where false negatives are critical, weighting the loss function to increase the penalty on false negatives helps the model better detect rare cases.
3. **Object Detection:** In autonomous driving, identifying pedestrians is more crucial than identifying static objects like signs. A weighted loss function can prioritize pedestrian detection to reduce the risk of accidents.
### Commonly Used Weighted Loss Functions:
- **Weighted Cross-Entropy:** Often applied in classification, where each class is assigned a weight based on its frequency or importance.
- **Focal Loss:** Assigns higher weight to hard-to-classify or rare examples, useful in scenarios with significant class imbalance.

In deep learning, the **choice of activation function** and loss function must align for the model to train effectively. Different activation functions shape the output of each layer, which can influence the type of loss function that best measures the error. 
# Here are some effective and problematic activation-loss function combinations:
### 1. Sigmoid Activation + Binary Cross-Entropy Loss
- **Effectiveness:** This is a classic combination for binary classification tasks. The sigmoid activation function compresses outputs between 0 and 1, which aligns well with Binary Cross-Entropy loss, allowing the model to interpret outputs as probabilities. This combination helps in minimizing log loss by penalizing incorrect high-confidence predictions.
- **Problematic Use:** Using sigmoid activation with Mean Squared Error (MSE) can lead to slow convergence, as MSE is less sensitive to large gradients and may result in vanishing gradient issues.

### 2. Softmax Activation + Categorical Cross-Entropy Loss
- **Effectiveness:** This combination is standard for multi-class classification. Softmax normalizes outputs into a probability distribution across multiple classes, making it a natural fit with Categorical Cross-Entropy loss, which measures the error in predicting one correct class out of multiple options.
- **Problematic Use:** Using Softmax with Binary Cross-Entropy can be problematic because Binary Cross-Entropy is intended for binary (two-class) outputs, whereas Softmax is designed for multi-class outputs.
### 3. Linear Activation + Mean Squared Error (MSE) or Mean Absolute Error (MAE)
- **Effectiveness:** This combination is ideal for regression tasks, where the target variable is continuous. Linear activation provides unrestricted output values, and MSE or MAE then measure the distance between predicted and actual values. MSE is more sensitive to outliers, while MAE is more robust for data with outliers.
- **Problematic Use:** Pairing a linear activation with Binary Cross-Entropy would be ineffective, as Binary Cross-Entropy expects probabilities (bounded between 0 and 1) and would not handle unrestricted outputs well.
### 4. ReLU Activation + Huber Loss
- **Effectiveness:** ReLU activation is commonly used in regression models with Huber Loss, especially when dealing with outliers. Huber Loss combines aspects of MSE and MAE, providing a balance between sensitivity to errors and robustness to outliers, which complements ReLU’s unrestricted positive output.
- **Problematic Use:** Using ReLU with Binary Cross-Entropy is inappropriate for binary classification, as ReLU outputs are not naturally bounded between 0 and 1, resulting in outputs that cannot represent probabilities.

## Summary:
Some recommended combinations include:
- Sigmoid + Binary Cross-Entropy for binary classification
- Softmax + Categorical Cross-Entropy for multi-class classification
- Linear + MSE/MAE for regression
- ReLU + Huber Loss for robust regression

In neural networks, **optimization** refers to the process of adjusting the model’s parameters (weights and biases) to minimize the loss function, which measures the difference between predicted and actual values. The aim is to find the set of parameters that results in the lowest possible loss, thereby improving the model’s accuracy on the task at hand.

# Why Optimizers Are Important
Optimizers are crucial in training neural networks because they dictate how and how fast the model’s parameters are updated during training. Effective optimizers:
1. **Control Convergence:** They determine the speed and stability of the model’s convergence towards the optimal or near-optimal solution.
2. **Navigate Complex Loss Landscapes:** Neural networks often have high-dimensional and non-convex loss surfaces. Optimizers help navigate these landscapes to avoid poor local minima and saddle points.
3. **Prevent Overfitting and Underfitting:** By adjusting learning rates and managing step sizes, optimizers can help achieve a balance between model accuracy and generalization.

# Common Optimizers
Some widely used optimizers include:
1. **Stochastic Gradient Descent (SGD):** Updates weights based on a random subset of data, improving training speed.
2. **Adam:** Adapts learning rates for each parameter, making it suitable for deep networks and achieving faster convergence.

In summary, optimizers play a key role in making training efficient, effective, and stable, enabling neural networks to reach high accuracy while avoiding issues like vanishing or exploding gradients.

Selecting the right optimizer for a deep learning task can be challenging, as different optimizers affect training dynamics, convergence speed, model performance, and stability in various ways. Here are some key challenges and how the choice of optimizer influences training:
# Challenges in Selecting an Optimizer
### 1. Model Complexity and Architecture:
- **Challenge:** Optimizers may work differently depending on the architecture (e.g., CNNs, RNNs, Transformers). Some optimizers, like Adam, are more adaptable across architectures, while others, like SGD, may need tuning to achieve good performance.
- **Impact:** If the optimizer isn’t suited to the model’s complexity, it may lead to slower convergence, suboptimal performance, or even failure to converge.

### 2. Convergence Speed and Stability:
- **Challenge:** Faster-converging optimizers, like Adam, can potentially reach a solution quicker, but may risk overshooting optimal solutions in complex landscapes. SGD with momentum can achieve stable convergence but may be slower.
- **Impact:** The right optimizer can balance convergence speed and stability, while an unsuitable one might cause oscillations, slow progress, or poor generalization.
  
### 3. Handling of Sparse vs. Dense Data:
- **Challenge:** Optimizers like AdaGrad and Adam perform well with sparse data by adapting learning rates for infrequently updated parameters, while standard SGD might struggle with these cases.
- **Impact:** Choosing an optimizer not tailored for the data type can lead to inefficient parameter updates, wasting computation, and potentially getting stuck in poor solutions.

### 4. Sensitivity to Hyperparameters:
- **Challenge:** Some optimizers require careful tuning of hyperparameters like the learning rate, momentum, or decay factors, which can be sensitive and difficult to get right for new tasks or datasets.
- **Impact:** Hyperparameters that aren’t well-suited to the optimizer can lead to unstable training, poor convergence, or even divergent loss.

### 5. Impact of Non-Stationary Data:
- **Challenge:** Tasks with data that changes over time or has inherent variability (e.g., RNNs for time-series forecasting) benefit from optimizers like RMSprop or Adam that adapt well to non-stationary data.
- **Impact:** Using an optimizer that can’t handle non-stationarity may result in oscillations or a failure to capture patterns effectively.

### 6. Computational Resources and Efficiency:
- **Challenge:** Optimizers like Adam require additional memory and computational overhead due to storing additional parameters (e.g., first and second moments), while SGD is less memory-intensive.
- **Impact:** The choice of optimizer affects training efficiency, and a computationally expensive optimizer might slow down training on large datasets or deep models.

# How Optimizer Choice Affects Training Dynamics and Convergence
- **Learning Rate Adjustment:** Adaptive optimizers like Adam and RMSprop adjust the learning rate dynamically, making them generally faster and more robust. This adjustment helps models converge more quickly and avoids getting stuck in local minima.
- **Gradient Oscillations:** Some optimizers, such as SGD without momentum, can lead to oscillations, especially in steep or noisy regions of the loss surface. Adding momentum helps smooth updates, resulting in a more stable convergence path.
- **Sensitivity to Saddle Points:** Optimizers like SGD are more prone to getting stuck in saddle points, while adaptive optimizers can often navigate around them, leading to smoother convergence.
- **Generalization:** Different optimizers affect how well the model generalizes to unseen data. SGD with momentum is known for achieving good generalization, as its slower convergence can avoid overfitting compared to faster optimizers like Adam, which can sometimes overfit if not carefully regularized.
Summary
Choosing the right optimizer requires considering the model architecture, data characteristics, computational constraints, and tuning requirements. Each optimizer influences the training dynamics, affecting convergence stability, speed, and final model performance

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adagrad
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
def create_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
    return model
```

```python
def train_and_evaluate(optimizer):
    model = create_model()
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    # Train the model
    history = model.fit(x_train, y_train, epochs=5, validation_split=0.2, verbose=0)
    
    # Evaluate on the test data
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    return history, test_loss, test_accuracy
```

```python
# List of optimizers to experiment with
optimizers = {
    'Adam': Adam(),
    'SGD': SGD(),
    'RMSprop': RMSprop(),
    'Adagrad': Adagrad()
}

results = {}
for name, optimizer in optimizers.items():
    history, test_loss, test_accuracy = train_and_evaluate(optimizer)
    results[name] = {
        'history': history,
        'test_loss': test_loss,
        'test_accuracy': test_accuracy
    }
    print(f"{name} - Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```

# <span style="color:blue; font-size:30px"> 6. Explore the role of momentum in optimization algorithms, such as SGD with momentum and Adam. How does momentum affect the optimization process, and under what circumstances might it be beneficial or detrimental?</span>

**Momentum** is a technique in optimization algorithms designed to accelerate the convergence of gradient descent by maintaining a velocity vector that aggregates past gradients. By incorporating this vector into the update step, momentum can help the optimizer move more smoothly toward the minimum, reducing oscillations in directions with steep gradients and speeding up convergence on flat surfaces.

### How Momentum Works
In optimizers like SGD with momentum, a fraction of the previous update is added to the current update. This addition creates a “momentum effect,” pushing the optimization in the desired direction with accumulated strength, making it easier to traverse narrow valleys in the loss surface and smoothing out noisy gradients. Momentum is particularly useful for navigating sharp or saddle regions in the loss landscape where gradients may oscillate.

In **Adam (which stands for Adaptive Moment Estimation)**, momentum is combined with adaptive learning rates. Adam uses exponential moving averages of both gradients and squared gradients, balancing momentum with element-wise scaling, which helps Adam adaptively handle non-stationary objective functions and sparse gradients.

### Benefits of Momentum
- **Accelerated Convergence:** Momentum helps reduce oscillations along steep dimensions, allowing for faster convergence along flat or narrow valleys.
- **Smoother Updates:** It reduces fluctuations in gradient descent, especially beneficial in high-dimensional spaces or noisy data.
- **Escaping Local Minima:** Momentum can help the optimizer escape shallow local minima by sustaining directional strength, pushing it toward a more promising solution.
### When Momentum Might Be Beneficial
- **Non-Convex Loss Landscapes:** Momentum is effective in non-convex optimization where the landscape has many saddle points and sharp gradients.
- **High-Dimensional Data:** It smooths the noisy updates often seen with high-dimensional data, making the training process more stable.
- **Sparse Gradients:** Optimizers like Adam, which use momentum with adaptive learning rates, handle sparse gradients well and can improve performance in tasks like NLP and image recognition.
### When Momentum Might Be Detrimental
- **Over-Acceleration:** If the momentum is too high, the optimizer may overshoot, missing the minimum or oscillating around it.
- **Sharp Minima:** Momentum can sometimes trap the model in sharp minima, which may generalize poorly on new data.

### Summary
Momentum accelerates convergence and stabilizes updates, especially beneficial in high-dimensional, noisy, or non-convex settings. However, it must be tuned carefully to avoid overshooting, particularly in optimizers like SGD with momentum.

# <span style="color:blue; font-size:30px">  7. Discuss the importance of hyperparameter tuning in optimizing deep learning models. How do hyperparameters, such as learning rate and momentum, interact with the choice of optimizer? Propose a systematic approach for hyperparameter tuning in the context of deep learning optimization.</span>

**Hyperparameter tuning** is essential in deep learning because hyperparameters, such as learning rate, momentum, batch size, and the choice of optimizer, directly impact model convergence, training stability, and final performance. These hyperparameters control various aspects of the optimization process, influencing how well a model learns from data. A carefully tuned set of hyperparameters can prevent issues like overfitting, underfitting, slow convergence, or even divergence during training.

## Key Hyperparameters and Their Interaction with Optimizers
- **Learning Rate:** Determines the step size in gradient descent. Too high a learning rate can cause the model to overshoot minima, while too low a rate - slows training. It needs to be tuned in concert with the optimizer; for instance, Adam can tolerate a slightly higher learning rate due to adaptive updates, while SGD often requires more careful tuning.

- **Momentum:** Helps accelerate convergence in gradient descent by adding a fraction of the previous update to the current step. High momentum can improve convergence speed, especially with SGD, but requires careful balancing with the learning rate to avoid overshooting.

- **Batch Size:** Impacts gradient estimates. Larger batch sizes give smoother updates but require more memory, whereas smaller batch sizes introduce more noise into the gradient, which can help the model escape local minima.

## Systematic Approach for Hyperparameter Tuning
### 1. Select a Search Method:

- **Grid Search:** Searches across a grid of hyperparameter values; effective for a small number of hyperparameters but computationally expensive for larger spaces.
- **Random Search:** Samples values randomly within a range, making it more efficient and often effective when only a few hyperparameters strongly influence performance.

# <span style="color:blue; font-size:30px"> 1. Explain the concept of forward propagation in a neural network.</span>

**Forward propagation**, also known as forward pass, is the process of computing the output of a neural network by feeding input data through the network’s layers in a sequential manner. It’s a crucial step in training a neural network, as it generates the predicted output based on the current weights and biases.

Here’s a step-by-step breakdown of forward propagation:

- **Input Layer:** The input data is fed into the network’s input layer, which consists of neurons (nodes) that receive the input values.
- **Hidden Layers:** The output from the input layer is propagated through one or more hidden layers, where each neuron applies an activation function (e.g., ReLU, Sigmoid, Tanh) to the weighted sum of its inputs. This process is repeated for each neuron in the hidden layer.
- **Output Layer:** The output from the hidden layers is fed into the output layer, where the final prediction is computed using a similar process as in the hidden layers.
- **Activation Functions:** At each layer, the output is passed through an activation function, which introduces non-linearity to the model, allowing it to learn complex relationships between inputs and outputs.
- **Weighted Sum:** The output from each neuron is computed by multiplying the input values by the corresponding weights and adding a bias term.
The forward propagation process can be mathematically represented as:

**output = activation(∑(weights \* inputs) + bias)**

where activation is the chosen activation function, weights are the learned weights, inputs are the input values, and bias is the bias term.

The output generated by forward propagation is then compared to the actual target output (in supervised learning) to calculate the error, which is used to update the weights and biases during the backpropagation phase.

# Chain Rule in Backpropagation
The chain rule plays a crucial role in backpropagation, a fundamental algorithm for training neural networks. Its primary purpose is to efficiently compute the gradients of the loss function with respect to the model’s parameters, enabling the optimization process to update the weights and biases.

# Key Application: 
### Computing Gradients
In backpropagation, the chain rule is used to compute the gradients of the loss function (e.g., mean squared error or cross-entropy) with respect to each weight and bias in the network. This is achieved by iteratively applying the chain rule to the composite functions that define the neural network’s architecture.

### Efficient Gradient Computation
The chain rule allows for an efficient computation of gradients by breaking down the complex, layered architecture of the neural network into smaller, more manageable components. This enables the algorithm to propagate errors (or gradients) backwards through the network, layer by layer, and compute the gradients of the loss function with respect to each parameter.

### Practical Significance
Without the chain rule, computing gradients would be a computationally expensive and impractical task, especially for deep neural networks. The chain rule’s application in backpropagation makes it possible to train large, complex models efficiently, which has been instrumental in the success of deep learning.


# <span style="color:blue; font-size:30px">5. Implement the forward propagation process for a simple neural network with one hidden layer using NumPy.</span>


Here is a Python implementation of forward propagation for a simple neural network with one hidden layer using NumPy:
# Neural Network Description:
- Input layer with n features.
- One hidden layer with h neurons using the ReLU activation function.
- Output layer with o neurons using the sigmoid activation function.

```python
import numpy as np
```

```python
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Initialize parameters for the network
def initialize_parameters(input_size, hidden_size, output_size):
    np.random.seed(42)  # For reproducibility
    W1 = np.random.randn(hidden_size, input_size) * 0.01  # Hidden layer weights
    b1 = np.zeros((hidden_size, 1))                      # Hidden layer biases
    W2 = np.random.randn(output_size, hidden_size) * 0.01  # Output layer weights
    b2 = np.zeros((output_size, 1))                       # Output layer biases
    return W1, b1, W2, b2

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    # Hidden layer computation
    Z1 = np.dot(W1, X) + b1  # Linear transformation
    A1 = relu(Z1)            # Activation function

    # Output layer computation
    Z2 = np.dot(W2, A1) + b2  # Linear transformation
    A2 = sigmoid(Z2)          # Activation function

    # Cache values for potential backpropagation
    cache = {
        "Z1": Z1, "A1": A1,
        "Z2": Z2, "A2": A2
    }

    return A2, cache


# Example usage
if __name__ == "__main__":
    # Define input size, hidden layer size, and output size
    input_size = 3
    hidden_size = 4
    output_size = 1

    # Initialize parameters
    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)

    # Define a sample input (3 features, 2 examples)
    X = np.array([[0.1, 0.2], [0.4, 0.5], [0.7, 0.8]])

    # Perform forward propagation
    A2, cache = forward_propagation(X, W1, b1, W2, b2)

    print("Output of the neural network (A2):")
    print(A2)
```
